###Room Type ve price

```python
plt.figure(figsize=(10, 6))
sns.set_theme(style='whitegrid')
sns.boxplot(y='room_type', x='price', hue='room_type', palette='Set3', linewidth=1.5, fliersize=1.5, data=df)

annotation_colors = ['red', 'green','purple', 'blue', 'orange', ]
colors = sns.color_palette('Set3', len(df['room_type'].unique()))
for i, room_type in enumerate(df['room_type'].unique()):
    median_price = df[df['room_type'] == room_type]['price'].median()
    color_index = i % len(annotation_colors)
    plt.text(df['price'].max() + 50, i, f'{room_type}: ${median_price}', ha='right', va='center', fontsize=12, color=annotation_colors[color_index], weight='bold', bbox=dict(facecolor='white', alpha=0.5, edgecolor=annotation_colors[color_index], boxstyle='round,pad=0.3'))

plt.title('Room Type vs. Price', fontsize=15)
plt.xlabel('Price', fontsize=12)
plt.ylabel('Room Type', fontsize=12)
plt.legend(title='Room Type')
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (8, 5)
ax= sns.countplot(y='room_type',data=df,palette='bright')

total = len(df['room_type'])
for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_width()/total)
        x = p.get_x() + p.get_width() + 0.02
        y = p.get_y() + p.get_height()/2
        ax.annotate(percentage, (x, y))

plt.title('Count of each room types in NYC')
plt.xlabel('Rooms')
plt.xticks(rotation=90)
plt.ylabel('Room Counts')

plt.show()
```
